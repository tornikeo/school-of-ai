# Metric spaces

#### Prove that:
For every metric space (X, d):
- For eny $\mathcal{U} = \{U | U$ is open in $(X, d)\}$ holds $\bigcup_{U \in \mathcal{U}} U$ is open in $(X, d)$
- For eny two $U, V \subset X$ open in $(X, d)$ holds: $U \cap V$ is open in $(X, d)$
- $X$ is open in $(X, d)$
- $\emptyset$ is open in $(X, d)$

#### proofs
- Let $x \in \bigcup_{i \in I}U_i$. Then the $x \in U_{j}$ for some $j \in I$. Therefore $x$ is an element of an open set and $\exists \epsilon \gt : B_{\epsilon}(x)$ and this open ball is a subset of $U_{j}$. and $U_{j} \subseteq \bigcup_{U\in \mathcal{U}} U_i$. Therefore, the union is also open.
- Let $x \in U \cap V$, then because $x \in V$, $\exists \epsilon_v : B_{\epsilon_v}(x) \subset V$, and same is true for some $\epsilon_u$ for $U$. now we choose the smaller $\epsilon$ that fits in both sets properly. Therefore there is always a ball around every element of intersection of $U$ and $V$ that falls within the set. By definition the resulting set is open
- Open set is, by definition, where every point inside of it has a ball with some radius that is a proper subset of the set. Let $x \in X$. The open ball of $x$ in $(X,d)$ is a subset of $X$, by definition. Therefore X is open in $(X,d)$.
- Empty set is open because every point in empty set is an interior point. This is true, because there are no points in it.

#### Prove that for set $U \subset X$ is open if and only if for each point $x \in U$ there exists the open neighbourhood $V$ of $x$ such that $V \subset U$

#### proof
- Let $U$ be open, then take any point $x \in U$. Consider a neighbourhood $V$ of $x$. For the sake of contradiction suppose that there is no $V$ such that it is a proper subset of $U$. That means, for any neighbourhood $V$ of $x$ there are always points in $V$ that are not in $U$. This makes $x$ by definition a boundary point. And thus $ x \notin U$, Which is a contradiction.
- Let $V = \bigcup_{x \in U} V_x$ be a union of all open neighbourhoods of every point $x \in U$. Consider the $V$. It must be open, as "Union of Open Sets of Metric Space is Open". And $\forall x : x\in U \iff x\in V$. So $U \equiv V$ and $U$ is open.

#### Prove that, closed ball is closed subset in $(X, d)$


#### proof
- We prove this by proving that complement of closed ball is open. Let $a \in B_{\epsilon}(x)$. Then by definition of closed ball, $d(x,a) > \epsilon$, denote $\delta := d(x,a) - \epsilon$, $\delta > 0$. Let $b \in B_{\delta}(a)$. Then $d(x,b) \geq d(x,a) - \delta = \epsilon $. Therefore, $b \notin B_{\epsilon}^c(x)$. So, there is an open neighbourhood for every point outside of $B_{\epsilon}^c(x)$. Therefore the closed ball is a closed subset.

- Same, $(A \cup B)^c = A^c \cap B^c$. Given that $A$ and $B$ are closed, Their complements are open and $A\cup B$ is closed.

Definition of Cauchy sequence: sequence $\left \langle {x_n} \right \rangle$ is a Cauchy sequence if and only if $$\forall \epsilon >0: \exists N \in \mathbb{N}: \forall m, n \in \mathbb{N}: m, n \ge N: d \left({x_n,x_m}\right) < \epsilon$$
From definition of limit, $$\forall \epsilon > 0 : \exists N \in \mathbb{N} : \forall n \in \mathbb{N} > N \to d(x_n, x) < \epsilon $$

Consider another sequence $x_m$ that has it's own $M$, such that whenever $m > M$ $d(x_m, x) < \epsilon / 2$
Now we combine the two limits, with the first limit being less than $ \epsilon / 2$ for the sake of simplicity.
$$ d(x_n,x) + d(x_m, x) < \epsilon $$
From triangle inequality
$$ \epsilon > d(x_n,x) + d(x, n_m) \ge d(x_n, x_m) $$
Therefore, completing the proof
